```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 

# Q3 & 4
df = pd.read_csv('Iowa_Liquor.csv', low_memory=False)
df.dropna()   #why isn't it dropping the NaN values???

df['Year'] = df['Date'].apply(lambda x: int(x.split('/')[2]))

df_2015 = df.loc[df['Year']==15]

print len(df_2015['Store Number'].unique())

def erase(symbol):
    symbol = symbol.replace('$','')
    symbol = symbol.replace(',','')
    return float(symbol)

df_2015['State Bottle Cost'] = df_2015['State Bottle Cost'].apply(lambda x: erase(x))
df_2015['State Bottle Retail'] = df_2015['State Bottle Retail'].apply(lambda x: erase(x))
df_2015['Sale (Dollars)'] = df_2015['Sale (Dollars)'].apply(lambda x: erase(x))

#Make a pivot table for sales in 2015 by store number

ts2015 = pd.pivot_table(df_2015, index = 'Store Number', values = ['Sale (Dollars)'], aggfunc = np.sum)
print ts2015

#this gives the total sales for each store in 2015

df_2015['At Cost'] = df_2015['State Bottle Cost'] * df_2015['Bottles Sold']

print df_2015.head()

df_2015['Store Profit'] = df_2015['Sale (Dollars)'] - df_2015['At Cost']

print df_2015['Store Profit']

profit2015 = pd.pivot_table(df_2015, index = 'Store Number', values = ['Store Profit'], aggfunc= np.sum)

print profit2015



#5 To make larger categories and a new column for those categories
def categorize(item):   
    try: 
        if 'VODKA' in item:
            item = 'VODKA'
        elif 'RUM' in item:
            item = 'RUM'
        elif 'GIN' in item:
            item = 'GIN'
        elif 'TEQUILA' in item:
            item = 'TEQUILA'
        elif 'SCHNAPPS' in item:
            item = 'SCHNAPPS'
        elif 'WHISK' in item or 'SCOTCH' in item or 'BOURBON' in item:
            item = 'WHISKY'
        elif 'BRAND' in item:
            item = 'BRANDY'
        else:
            item = 'OTHER'
        return item 
    except:
        pass

df['Liquor Types'] = df['Category Name'].apply(categorize)
df.head()

#to see the number of total liquors in each liquor type
df['Liquor Types'].value_counts()

#to see the number of different liquors in each liquor type
#WHISKY:
df.loc[df['Liquor Types'] == 'WHISKY', 'Category Name'].unique()
df.loc[df['Liquor Types'] == 'WHISKY', 'Category Name'].value_counts()
#VODKA
df.loc[df['Liquor Types'] == 'VODKA', 'Category Name'].unique()
df.loc[df['Liquor Types'] == 'VODKA', 'Category Name'].value_counts()
#GIN
df.loc[df['Liquor Types'] == 'GIN', 'Category Name'].unique()
df.loc[df['Liquor Types'] == 'GIN', 'Category Name'].value_counts()
#RUM
df.loc[df['Liquor Types'] == 'RUM', 'Category Name'].unique()
df.loc[df['Liquor Types'] == 'RUM', 'Category Name'].value_counts()
#TEQUILA
df.loc[df['Liquor Types'] == 'TEQUILA', 'Category Name'].unique()
df.loc[df['Liquor Types'] == 'TEQUILA', 'Category Name'].value_counts()
#BRANDY
df.loc[df['Liquor Types'] == 'BRANDY', 'Category Name'].unique()
df.loc[df['Liquor Types'] == 'BRANDY', 'Category Name'].value_counts()
#SCHNAPPS
df.loc[df['Liquor Types'] == 'SCHNAPPS', 'Category Name'].unique()
df.loc[df['Liquor Types'] == 'SCHNAPPS', 'Category Name'].value_counts()
#OTHER
df.loc[df['Liquor Types'] == 'OTHER', 'Category Name'].unique()
df.loc[df['Liquor Types'] == 'OTHER', 'Category Name'].value_counts()


#to find the number of bottle sold for each broad category... another pivot table..

Bottles_Sold_Per_Category = pd.pivot_table(df, index= 'Liquor Types', values= ['Bottles Sold'], aggfunc= np.sum)
print Bottles_Sold_Per_Category

df.head()
#to find the profit for each broad category - have to drop $ from the rest of the dataframe (only dropped from 2015 before)

def erase(symbol):
    symbol = symbol.replace('$','')
    symbol = symbol.replace(',','')
    return float(symbol)

df['State Bottle Cost'] = df['State Bottle Cost'].apply(lambda x: erase(x))
df['State Bottle Retail'] = df['State Bottle Retail'].apply(lambda x: erase(x))
df['Sale (Dollars)'] = df['Sale (Dollars)'].apply(lambda x: erase(x))

#double checking
df.head()

#Creating a profits per bottle column
df['Profit Per Bottle'] = df['State Bottle Retail'] - df['State Bottle Cost']

#Total profits column
df['Total Profit'] = df['Profit Per Bottle'] * df['Bottles Sold']
#Whisky had the most bottles sold 


#Pivot table for profits for each category

Profits_By_Category = pd.pivot_table(df, index= 'Liquor Types', values= ['Total Profit'], aggfunc= np.sum)
print Profits_By_Category
#I want to limit the number of places after the decimal too, but I'm not sure how to do that.
#Whisky was the most profitable



#6 Pivot table with Broad Category, liquors in each category, max, min, volume price per bottle:

table = pd.pivot_table(df, index=['Liquor Types','Category Name'], values=['State Bottle Retail','Volume Sold (Liters)'], 
                       aggfunc=[np.max, np.min, np.sum, np.mean])
print table



#7 County by county differences
#Different counties:

df['County'].value_counts()

by_county1 = df.groupby(['County', 'Liquor Types'])['Liquor Types'].count()
by_county1

by_county2 = pd.pivot_table(df, index=['County', 'Liquor Types'], values=['Volume Sold (Liters)'], aggfunc=np.sum)
by_county2
#wasn't sure whether to use a pivot table or a groupby... couldn't figure out how to sort the values to rank them



#8

#Populations by County....

len(df['County'].value_counts())

# 5 smallest counties: Adams, Ringgold, Audubon, Taylor, Wayne
# 5 largest counties: Polk, Linn, Scott, Black Hawk, Johnson

def size(county):
    if 'Polk' in county: 
        county = 'HighPop'
    elif 'Linn' in county:
        county = 'HighPop'
    elif 'Scott' in county: 
        county = 'HighPop'
    elif 'Black Hawk' in county: 
        county = 'HighPop'
    elif 'Johnson' in county:
        county = 'HighPop'
    elif 'Adams' in county: 
        county = 'LowPop'
    elif 'Ringgold' in county:
        county = 'LowPop'
    elif 'Audubon' in county:
        county = 'LowPop'
    elif 'Taylor' in county: 
        county = 'LowPop'
    elif 'Wayne' in county:
        county = "LowPop"
    else:
        pass

High_Pop = df.loc[(df['County Name'] == 'HighPop')]
Low_Pop = df.loc[(df['County Name'] == 'LowPop')]

#I ran out of time and this is as far as I got with this one. Will continue to think on how to proceed. 
#I could probably do a groupby with the county size and store number, then find the amount of unique store numbers per county. Compare totals. I could do the same with # of bottles purchased. 



#9
# Plotting

#again, I ran out of time, but I would choose number 4 to plot. I'd make a stacked bar chart. On the x-axis I'd have different stores on the on the y axis I'd have $. 
# One segment of the stacked bars would show the total sales for the store in 2015 (would ultimately be the upper segment of the bar). The lower segment of the bar would indicate the stores profit in 2015.












```
